```python
import numpy as np
import scipy.io
from libsvm.svmutil import *
from sklearn.metrics import accuracy_score
```

```python
train_data=scipy.io.loadmat("./sample_data/trainData.mat")
```

```python
test_data=scipy.io.loadmat("./sample_data/testData.mat")
```

```python
train_data['X1'].shape
```

```python
train_data_X1=train_data["X1"]
train_data_X2=train_data["X2"]
train_data_X3=train_data["X3"]
train_data_Y=train_data['Y'].reshape(len(train_data['Y']),)

test_data_X1=test_data["X1"]
test_data_X2=test_data["X2"]
test_data_X3=test_data["X3"]
test_data_Y=test_data['Y'].reshape(len(test_data['Y']),)
```

```python
test_data_Y.shape
```

```python
y, x = np.array([1,-1]), np.array([[1,0,1], [-1,0,-1]])

# print(np.array([1,0,1], [-1,0,-1]))
```

```python
prob=svm_problem(train_data_Y, train_data_X1)
param = svm_parameter('-c 10 -t 0')
m = svm_train(prob, param)
# m1= svm_train(train_data_Y, train_data_X1, '-c 10 -t 0')
# m3= svm_train(train_data_Y, train_data_X3, '-c 10 -t 0')
p_label, p_acc, p_val = svm_predict(test_data_Y,test_data_X1, m)
```

```python
prob_1=svm_problem(train_data_Y, train_data_X1)
param_1 = svm_parameter('-c 10 -t 0 -b 1')
m_1 = svm_train(prob_1, param_1)
p_label_1, p_acc_1, p_val_1  = svm_predict(test_data_Y,test_data_X1, m_1,"-b 1")
```

```python
prob_2=svm_problem(train_data_Y, train_data_X2)
param_2 = svm_parameter('-c 10 -t 0')
m_2 = svm_train(prob_2, param_2)
p_label_2, p_acc_2, p_val_2 = svm_predict(test_data_Y,test_data_X2, m_2)
```

```python
prob_3=svm_problem(train_data_Y, train_data_X2)
param_3 = svm_parameter('-c 10 -t 0 -b 1')
m_3 = svm_train(prob_3, param_3)
p_label_3, p_acc_3, p_val_3 = svm_predict(test_data_Y,test_data_X2, m,'-b 1')
```

```python
prob_4=svm_problem(train_data_Y, train_data_X3)
param_4 = svm_parameter('-c 10 -t 0')
m_4 = svm_train(prob_4, param_4)
p_label_4, p_acc_4, p_val_4 = svm_predict(test_data_Y,test_data_X3, m_4)
```

```python
prob_5=svm_problem(train_data_Y, train_data_X3)
param_5 = svm_parameter('-c 10 -t 0 -b 1')
m_5 = svm_train(prob_5, param_5)
p_label_5, p_acc_5, p_val_5 = svm_predict(test_data_Y,test_data_X3, m_5,"-b 1")
```

```python
posterior_val=p_val_1+p_val_3+p_val_5
```
